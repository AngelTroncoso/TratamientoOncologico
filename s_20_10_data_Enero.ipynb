```python
dataset.groupby('TUMOR PRIMARIO')['TAMAÑO'].aggregate(['min','max','mean']) \
        .rename(columns={'min':'Minimo', 'max':'Maximo', 'mean':'Promedio'}) \
        .sort_values(by='Maximo', ascending=False)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(16,6))
```

```python
fig = dataset.groupby('TUMOR PRIMARIO')['TAMAÑO'].mean().plot.bar(color='green')
fig.set_title('Tamaño Promedio del Tumor Primario', {'fontsize':22})
fig.set_ylabel('Tamaño Promedio del Tumor Primario ')
```
